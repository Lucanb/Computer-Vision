```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

a = 'http://profs.info.uaic.ro/~ancai/VA/lab/lena.tif'

image = cv.imread('lena.tif')
image = cv.cvtColor(image,cv.COLOR_BGR2RGB)
print(image.shape)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
image = cv.imread('lena.tif')
kernel2 = np.ones((5, 5), np.float32) / 25
blurred_img = cv.filter2D(src=image, ddepth=-1, kernel=kernel2)

sharpening_kernel = np.array([[0, -1, 0],
                              [-1, 5,-1],
                              [0, -1, 0]], np.float32)
sharpened_img = cv.filter2D(src=image, ddepth=-1, kernel=sharpening_kernel)

image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
blurred_img_rgb = cv.cvtColor(blurred_img, cv.COLOR_BGR2RGB)
sharpened_img_rgb = cv.cvtColor(sharpened_img, cv.COLOR_BGR2RGB)

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(blurred_img_rgb)
plt.title('Blurred Image (5x5 Kernel)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sharpened_img_rgb)
plt.title('Sharpened Image')
plt.axis('off')

plt.show()
```

```python
filter = np.array([[0, -2, 0],
                    [-2, 8,-2],
                    [0, -2, 0]], np.float32)
image = cv.imread('lena.tif')
new_image = cv.filter2D(src=image,ddepth=-1,kernel=filter)
imgRGB = cv.cvtColor(image,cv.COLOR_BGR2RGB)
newRGB_image = cv.cvtColor(new_image,cv.COLOR_BGR2RGB)
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(imgRGB)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(newRGB_image)
plt.title('New filter image')
plt.axis('off')
```

```python
def rotate_image(image, angle, clockwise=True): #trebuie Revizuit ------------
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    if not clockwise:
        angle = -angle
    M = cv.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv.warpAffine(image, M, (w, h))
    return rotated   #implementez folosind coordonate polare zic si 

image = cv.imread('lena.tif')

rotated_clockwise_45 = rotate_image(image, 45, clockwise=True)
rotated_counterclockwise_90 = rotate_image(image, 90, clockwise=False)

image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
rotated_clockwise_45_rgb = cv.cvtColor(rotated_clockwise_45, cv.COLOR_BGR2RGB)
rotated_counterclockwise_90_rgb = cv.cvtColor(rotated_counterclockwise_90, cv.COLOR_BGR2RGB)

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(rotated_clockwise_45_rgb)
plt.title('Rotated Clockwise (45)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(rotated_counterclockwise_90_rgb)
plt.title('Rotated Counterclockwise (90)')
plt.axis('off')

plt.show()
```

```python
def crop_image(image, x, y, width, height):
    cropped_image = image[y:y+height, x:x+width]
    return cropped_image

image = cv.imread('lena.tif')

cropped_img = crop_image(image, x=50, y=50, width=200, height=150)

image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
cropped_img_rgb = cv.cvtColor(cropped_img, cv.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cropped_img_rgb)
plt.title('Cropped Image')
plt.axis('off')

plt.show()
```
